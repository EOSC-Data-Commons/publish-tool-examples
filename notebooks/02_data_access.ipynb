# Data Access: Searching and Retrieving Datasets

This notebook demonstrates how to search the EOSC Data Commons catalogue and retrieve datasets for use in your research workflows.

## 1. Setup

```python
import os
import json
import requests
import pandas as pd

API_TOKEN = os.environ.get("EOSC_API_TOKEN", "")
API_BASE_URL = os.environ.get("EOSC_API_BASE_URL", "https://api.eosc-portal.eu")

HEADERS = {
    "Authorization": f"Bearer {API_TOKEN}",
    "Content-Type": "application/json",
    "Accept": "application/json",
}
```

## 2. Search the Catalogue

Use the search endpoint to find datasets by keyword, subject, or other metadata fields.

```python
def search_datasets(query: str, page: int = 1, page_size: int = 10) -> dict:
    """
    Search the EOSC catalogue for datasets matching the given query.

    Parameters
    ----------
    query : str
        Free-text search query.
    page : int
        Page number (1-indexed).
    page_size : int
        Number of results per page.

    Returns
    -------
    dict
        Parsed JSON response from the API.
    """
    url = f"{API_BASE_URL}/datasets"
    params = {"q": query, "page": page, "pageSize": page_size}
    response = requests.get(url, headers=HEADERS, params=params, timeout=30)
    response.raise_for_status()
    return response.json()


# Example: search for climate datasets
# results = search_datasets("climate")
# print(json.dumps(results, indent=2))
```

## 3. Retrieve Dataset Metadata

Fetch detailed metadata for a specific dataset using its identifier.

```python
def get_dataset_metadata(dataset_id: str) -> dict:
    """
    Retrieve full metadata for a dataset by its identifier.

    Parameters
    ----------
    dataset_id : str
        Unique identifier of the dataset.

    Returns
    -------
    dict
        Dataset metadata as a dictionary.
    """
    url = f"{API_BASE_URL}/datasets/{dataset_id}"
    response = requests.get(url, headers=HEADERS, timeout=30)
    response.raise_for_status()
    return response.json()


# Example usage:
# metadata = get_dataset_metadata("my-dataset-id")
# print(json.dumps(metadata, indent=2))
```

## 4. Download a Dataset File

Download a specific file from a dataset to a local path.

```python
def download_file(file_url: str, local_path: str) -> None:
    """
    Download a file from the given URL to a local path.

    Parameters
    ----------
    file_url : str
        Direct download URL for the file.
    local_path : str
        Local filesystem path where the file will be saved.
    """
    response = requests.get(file_url, headers=HEADERS, stream=True, timeout=60)
    response.raise_for_status()
    with open(local_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"Downloaded file to: {local_path}")


# Example usage:
# download_file("https://example.com/dataset/file.csv", "./data/file.csv")
```
